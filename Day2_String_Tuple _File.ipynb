```python
#Project1:文字分析器
sentence = "Data Science is amazing and Data is powerful"
words = sentence.lower().split()
print(f"總共有{len(words)}個單字")

max_len = 0
longest = []
for w in words:
    if len(w) > max_len:
        max_len = len(w)
        longest = [w]
    
    elif len(w) == max_len:
        longest.append(w)


with open("project2.txt", "w", encoding="utf-8") as f:
    f.write("最長單字："+", ".join(longest))

with open("project2.txt", "r", encoding="utf-8") as f:
    content = f.read()
print(content)
```

```python
#Project 3:進階挑戰
raw = " Python, JAVA, C++, Python, java "
raw = raw.strip().lower().replace(",","").split()
print(raw)
print(f"共有{len(raw)}個單字")

max_len = 0
longest = []
for r in raw:
    if len(r) > max_len:
        max_len = len(r)
        longest = [r]
    elif len(r) == max_len:
        longest.append(r)
print("最長單字:"+",".join(set(longest))) #set() 重複的單字只會記錄一次

with open("language.txt","w", encoding="utf-8") as f:
    f.write(f"共有{len(raw)}個單字\n")
    f.write("最長單字："+", ".join(set(longest)))
```

```python
#更進階的 Python 寫法（Day 3 會學）
raw = " Python, JAVA, C++, Python, java "
words = raw.strip().lower().replace(",", "").split()

max_len = max(len(word) for word in words)
longest = [word for word in words if len(word) == max_len]

print(f"共有 {len(words)} 個單字")
print("最長單字：", ", ".join(set(longest)))
```
