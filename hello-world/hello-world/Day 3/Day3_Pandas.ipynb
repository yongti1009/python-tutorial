# Part 7：NumPy array 和 list 的差別

```python
#list: a+b 輸出是合併
a = [1,2,3]
b = [4,5,6]

print(a+b)

#Numpy: a+b 輸出是逐個元素相加
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a+b)
```

## Project 1：成績分析器 ★

```python
scores = np.array([78,85,90,92,88,76,95])

print(f"平均分數:{scores.mean():.1f}\n最高分:{scores.max()}\n最低分:{scores.min()}\n總分:{scores.sum()}\n標準差:{scores.std():.1f}")
```

## Project 2：BMI 分析器 ★

```python
import numpy as np

height = np.array([170, 165, 180, 175])
weight = np.array([65, 55, 80, 70])

BMI = weight / (height/100)**2
print(BMI)

#小數點後兩位(法1)
print(np.round(BMI, 2)) 
#法2
#for x in BMI:
    #print(f"{x:.2f}")

print(f"平均 BMI:{BMI.mean():.2f}\n最高 BMI:{BMI.max():.2f}\n最低 BMI:{BMI.min():.2f}")

```

## Project 4：公司銷售分析 ★★

```python
sales = np.array([120,150,180,200,170,220,250])
print(f"總銷售:{sales.sum():.1f}\n平均銷售:{sales.mean():.1f}\n最高銷售:{sales.max()}\n最低銷售:{sales.min()}")

print(f"高於平均的銷售:{sales[sales > sales.mean()]}") #Boolean Indexing [篩選條件]
```

### Boolean Indexing = 用 True / False 篩選資料。

import numpy as np

scores = np.array([78,85,92,60,55,88,95,70])

#哪些成績大於 80？
print (scores > 80)

得到:

[False  True  True False False  True  True False]

意思是：
| 分數 | >80?  |
| -- | ----- |
| 78 | False |
| 85 | True  |
| 92 | True  |
| 60 | False |
| 55 | False |
| 88 | True  |
| 95 | True  |
| 70 | False |

把剛剛的條件放進 []：

scores[scores > 80]

得到:

[85 92 88 95]


## Project 5：考試成績分析 ★★

```python
scores = np.array([78,85,92,60,55,88,95,70])
print(f"全班平均:{scores.mean():.1f}")
print(f"不及格成績:{scores[scores < 60]}")
print(f"不及格人數:{len(scores[scores<60])}")
```
