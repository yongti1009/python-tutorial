```python
#Project 7：計算平均函式
scores = [80, 90, 95, 88, 100]
scores.append(67)

def average(scores):
    total = 0
    for score in scores:
        total += score
    return total / len (scores)
print(f"Average = {average(scores):.1f}")
```

```python
#Project 8：文字統計器
#split() → 拆單字
#len(list) → 幾個單字
#len(string) → 幾個字元
#for + len() → 找最大值

sentence = "Data Science is awesome"

#單字切開 sentence.split()
words = sentence.split()  #['Data', 'Science', 'is', 'awesome']

#計算字串長度
print(f"共有{len (sentence)}個字元") 

#1 找出最長單字：用 for 逐個比較
longest = "" 
for w in words:
    if len(w) > len(longest):
        longest = w

#2 Python 直接幫你找最大長度
longest = max(words, key=len)

#3 若最多字源不只一個，也要全部輸出
    # 只用一個變數 → 只能抓一個最大值 
    # 用list → 才能抓「所有最大值」
max_len = 0
longest = []

for w in words:
    if len(w) > max_len:
        max_len = len(w) #不斷更新標準
        longest = [w]   # 重新開始
    elif len(w) == max_len:
        longest.append(w)

#4 進階用法
max_len = max(len(w) for w in words)
longest = [w for w in words if len(w) == max_len]

#輸出字元最多的單字
print(f"最長單字為{longest}")
#計算共有幾個單字
print(f"共有{len(words)}個單字") 
```

```python
#10 Project 10：迷你 Data Science 專案 
heights = [170, 165, 180, 175, 172]
weights = [66, 58,80,70,68]
person_BMI = [] #建立空的list(串列)
person = 0

#一對一資料 → 用 zip() 
# #zip(heights, weights)
for h, w in zip(heights, weights): 
    person += 1
    bmi = w/(h/100)**2
    person_BMI.append(bmi)
    print(f"Pearson {person} BMI = {bmi:.1f}")

print(f"平均 BMI {sum(person_BMI)/len(person_BMI):.1f}")
print(f"最高 BMI {max(person_BMI):.1f}")
print(f"最低 BMI {min(person_BMI):.1f}")

```

```python
#更專業寫法
heights = [170, 165, 180, 175, 172]
weights = [66, 58,80,70,68]
bmis = [w/(h/100)**2 for h, w in zip(heights, weights)]

for i, bmi in enumerate(bmis, 1):
    print(f"Person {i} BMI = {bmi:.1f}")

print("平均 BMI =", f"{sum(bmis)/len(bmis):.1f}")
print("最高 BMI =", f"{max(bmis):.1f}")
print("最低 BMI =", f"{min(bmis):.1f}")
```
