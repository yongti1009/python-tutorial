```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt

df = sns.load_dataset("titanic")
```

## Part 1：Histogram（直方圖）⭐⭐⭐⭐⭐

用來看「分布」

    plt.hist(df["age"].dropna())
    plt.show()

👉 觀察：

年齡分布

哪個年齡最多 


```python
#畫出：fare（票價）的分布圖

plt.hist(df["fare"].dropna())
plt.show()
```

```python
#畫出：age 分布圖，並觀察是否有偏態（skewed）
plt.hist(df["age"].dropna())
plt.show()
print(f"{df["age"].skew():.2f}")

```

0.39 (>0)表示輕微右偏

```python
#檢驗偏態
sns.boxplot(x=df["age"])
#右邊很多離值群
```
