```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("titanic")
```

## Part 1：groupby() ⭐⭐⭐⭐⭐
groupby 是 Data Scientist 最常用的功能之一。

```python
#依照性別分組在計算年齡平均
df.groupby("sex")["age"].mean()
```

```python
#依照艙等分組計算平均票價
df.groupby("class")["fare"].mean()
```

```python
#依照性別計算生還率
df.groupby("sex")["survived"].mean()
```

```python
#依照艙等計算平均年齡
df.groupby("class")["age"].mean()
```

```python
#依照embark_town計算平均價格
df.groupby("embark_town")["fare"].mean()
```

### Part 2：多個 groupby ⭐⭐⭐⭐⭐
可以同時依照兩個欄位分組。

```python
#不同性別 × 艙等的平均年齡。
df.groupby(["sex","class"])["age"].mean()
```

```python
#不同性別 × 艙等的生還率。
df.groupby(["sex", "class"])["survived"].mean()
```

### Part 3：pivot_table() ⭐⭐⭐⭐⭐
pivot table 可以把 groupby 的結果變成表格。

```python
pd.pivot_table(
    df,
    values = "survived",
    index = "sex",
    columns = "class"
)
```

```python
#建立不同性別 × 艙等的平均年齡表。
pd.pivot_table(
    df,
    values = "age",
    index = "sex",
    columns = "class"
)
```

```python
#建立同性別 × 艙等的平均票價表。
pd.pivot_table(
    df,
    values = "fare",
    index = "sex",
    columns = "class"
)
```

```python
#建立不同 embark_town × class 的平均生還率。
pd.pivot_table(
    df,
    values = "survived",
    index = "embark_town",
    columns = "class"
)
```

### Part 4：缺失值（Missing Value）⭐⭐⭐⭐⭐

```python
#先查看缺失值
df.isnull()
#統計每隔欄位缺失值數量
df.isnull().sum()
```

### Part 5：dropna() ⭐⭐⭐⭐⭐

```python
#刪除缺失值
df.dropna()
```

```python
#刪除age缺失值的新DataFrame
df2 = df.dropna(subset = ["age"])
```

```python
#比較刪除前與刪除後的資料量
print(len(df))
print(len(df2))
```

### Part 6：fillna() ⭐⭐⭐⭐⭐

```python
#使用平均值填補並建立新欄位
df["age_fill"] = df["age"].fillna(
    df["age"].mean()
)
```

```python
#利用平均票價填補缺失值。
df["fare_fill"] = df["fare"].fillna(
    df["fare"].mean()
)
```

```python
#確認新欄位缺失值數量
df["age_fill"].isnull().sum()
```

### Part 7：merge() ⭐⭐⭐⭐⭐
Data Scientist 幾乎每天都在合併資料。

```python
student = pd.DataFrame({
    "ID" : [1,2,3],
    "Name" : ["Amy", "Bob", "Cindy"]
})

score = pd.DataFrame({
    "ID" : [1,2,3],
    "Score" : [90,80,95]
})
#合併兩個dictionary
pd.merge(student, score)
```

```python
fruit = pd.DataFrame({
    "ID" : [1,2,3],
    "Product" : ["Apple", "Banana", "Orange" ]
})

price = pd.DataFrame({
    "ID" : [1,2,3],
    "Price" : [30, 20, 25]
})
pd.merge(fruit, price)
```

```python
name = pd.DataFrame({
    "ID" : [101,102,103],
    "Name" : ["Tom", "Mary", "John"]
})

depart = pd.DataFrame({
    "ID" : [101,102,103],
    "Department" : ["HR", "Finance", "IT"]
})

pd.merge(name, depart)
```

### Bonus Project ⭐⭐⭐⭐⭐

「女性是否比男性更容易生還？」

「First Class 是否比 Third Class 更容易生還？」

```python
sex_survival = df.groupby("sex")["survived"].mean()
sex_survival.sort_values(ascending= False) #在新的表格中排序
```

```python
class_survival = df.groupby("class")["survived"].mean()
class_survival.sort_values(ascending= False)
```

## 女性是否比男性更容易生還？ 

女生的平均生還率有74%，遠高於男生的19%，女生比男生更容易生還，因此可推測女性乘客在鐵達尼號事故中明顯較容易獲得救援機會。

## First Class 是否比 Third Class 更容易生還？

Fitrst Class的平均生還率有63%，Third Class的平均生還率僅24%，因此可推知在艙等越高的旅客，生還率也越高。

## Final Findings

本次分析發現：

1.女性的生還率顯著高於男性。

2.艙等越高，生還率越高。

3.First Class 為生還率最高的艙等。

4.Third Class 為生還率最低的艙等。

## 綜合結果顯示：

性別與艙等都與 Titanic 乘客的生還機率具有明顯關聯性。
