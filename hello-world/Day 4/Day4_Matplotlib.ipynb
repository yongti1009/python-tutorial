### Project 9：生還率分析 ⭐⭐⭐⭐

```python
df["survived"].value_counts().rename(index={0 :"Dead", 1 : "Survive"})

生還率 = 342/(549+342)*100 #等於平均值(不要把數字寫死)
print(生還率,"%")
```

loc

Pandas 設計者覺得：

用標籤(label)切片時，包含結尾比較直覺

所以：

df.loc[0:2]

會取：

0
1
2

```python
df.loc[0:2]
```

| 功能              | loc       | iloc         |
| --------------- | --------- | ------------ |
| 用什麼取值           | 標籤(label) | 位置(position) |
| 取單一列 (第二列)     | `loc[3]`  | `iloc[3]`    |
| 切片是否包含結尾        | ✅ 包含      | ❌ 不包含        |
| 遵守 Python slice | ❌         | ✅            |
